## 데이터 제외하기 dropna

```python
import pandas as pd
df = pd.read_csv('score.csv')
df
```

```python
df.dropna()# 전체 데이터 중에서 NaN을 포함하는 데이터 삭제(row 전체)
```

```python
df
```

- axis: index or columns
- how: any or all

```python
df.dropna(axis='index',how='any')#NaN 데이터가 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns')
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns',how='all')#해당 컬럼 전체가 NaN인 경우에만 Column 삭제
```
