# 데이터 선택(조건)
조건에 해당하는 데이터를 선택

```python
import pandas as pd
df = pd.read_csv('score.csv')
df
```

```python
df['키'] >= 185
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] #filt를 역으로 적용
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키']>=185, '수학'] #키가 185이상인 학생들의 수학 데이터
```

```python
df.loc[df['키'] >= 185, ['이름', '수학','과학']]
```

## 다양한 조건

### &(And)

```python
df.loc[(df['키']>=185) & (df['학교']=='북산고')]
```

### |(OR)

```python
df.loc[(df['키'] < 170)| (df['키'] > 200)]
```

## STR함수

```python
filt = df['이름'].str.startswith('송') #'송'씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태')# 이름에 '태'가 들어가는 사람
df[filt]
```

```python
df[~filt]
```

```python
langs=['PYTHON','Java']
filt = df['SW특기'].isin(langs) #SW특기가 Python이거나 Java인 사람
df[filt]
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df["SW특기"].str.contains('Java',na=True) #NaN데이터에 대해서 True로 설정
df[filt]
```

```python
filt = df["SW특기"].str.contains('Java',na=False)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java',na=False)
df[filt]
```
