# 데이터 정렬

```python
import pandas as pd
df = pd.read_csv('score.csv')
df
```

```python
df.sort_values('키')
```

```python
df.sort_values('키',ascending=False)
```

```python
df.sort_values(['수학','영어'],ascending=False)
```

```python
df.sort_values(['수학','영어'])
```

```python
df.sort_values(['수학','영어'],ascending=[True, False]) #수학은 오름차순, 영어는 내림차순으로 정렬
```

```python
df['키'].sort_values()
```

```python
df['키'].sort_values(ascending=False)
```
