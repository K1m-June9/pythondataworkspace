# Pandas
파이썬에서 사용하는 데이터 분석 라이브러리

```python
import pandas as pd
```

# 1.Series
1차원 데이터(정수, 실수, 문자열 등)

## Series 객체 생성
ex) 1월부터 4월까지 평균 온도 데이터(-20, -10, 10, 20)

```python
temp = pd.Series([-20,-10,10,20])
temp
```

```python
temp[0] #1월 온도
```

```python
temp[2]
```

# Series 객체 생성(Index 지정)

```python
temp = pd.Series([-20, -10, 10, 20], index = ['Jan','Feb','Mar','Apr'])
temp
```

```python
temp['Jan'] #index Jan에 해당하는 데이터 출력
```
