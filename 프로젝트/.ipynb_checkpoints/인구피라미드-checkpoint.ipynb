# 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
filename='202312_202312_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(filename,skiprows=3,index_col='행정기관',usecols='B,E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int) #','가 있으면 str로 인식할 수 있기 때문에 삭제 후 int형으로 변환
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(filename,skiprows=3,index_col='행정기관',usecols='B,AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns #컬럼명 통일
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int) #','가 있으면 str로 인식할 수 있기 때문에 삭제 후 int형으로 변환
df_w
```

# 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['font.family']='Malgun Gothic'--> Window
matplotlib.rcParams['font.family']='AppleGothic'#Mac
#matplotlib.rcParams['font.size']=10 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False#한글폰트 사용 시 -값 깨짐 방지
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0]//1000) #단위: 천 명
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.title("2023 대한민국 인구 피라미드")
plt.savefig('2023_대한민국 인구 피라미드.png',dpi=100)

plt.show()
```
