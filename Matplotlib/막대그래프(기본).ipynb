# 막대그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['font.family']='Malgun Gothic'--> Window
matplotlib.rcParams['font.family']='AppleGothic'#Mac
#matplotlib.rcParams['font.size']=10 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False#한글폰트 사용 시 -값 깨짐 방지
```

```python
labels=['강백호','서태웅','정대만']
values=[190,187,184]

plt.bar(labels,values)
```

```python
labels=['강백호','서태웅','정대만']
values=[190,187,184]

plt.bar(labels,values,color='r')
```

```python
labels=['강백호','서태웅','정대만']
values=[190,187,184]
color=['r','g','b']

plt.bar(labels,values,color=color,alpha=0.5)
```

```python
labels=['강백호','서태웅','정대만']
values=[190,187,184]

plt.bar(labels,values)
plt.ylim(175,195)
```

```python
plt.bar(labels,values,width=0.4)
```

```python
plt.bar(labels,values,width=0.3)
plt.xticks(rotation=45) #x축의 데이터를 45도로 설정
plt.yticks(rotation=45)# y축의 데이터를 45도로 설정
```

```python
labels=['강백호','서태웅','정대만']
values=[190,187,184]
ticks=['1번','2번','3번']

plt.bar(labels, values)
plt.xticks(labels,ticks)
```
