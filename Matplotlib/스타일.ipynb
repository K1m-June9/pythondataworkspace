# 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['font.family']='Malgun Gothic'--> Window
matplotlib.rcParams['font.family']='AppleGothic'#Mac
#matplotlib.rcParams['font.size']=10 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False#한글폰트 사용 시 -값 깨짐 방지
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,linewidth=5)
```

# Marker

```python
plt.plot(x,y,marker='o')
```

```python
plt.plot(x,y,marker='o',linestyle='None')
```

```python
plt.plot(x,y,marker='v')
```

```python
plt.plot(x,y,marker='v',markersize=10)
```

```python
plt.plot(x,y,marker='X',markersize=10)#더 많은 마커는 matplotlib marker 검색해서 찾아보셈
```

```python
plt.plot(x,y,marker='X',markersize=10,markeredgecolor='red')
```

```python
plt.plot(x,y,marker='o',markersize=10,markeredgecolor='red',markerfacecolor='yellow')
```

# Line Style

```python
plt.plot(x,y,linestyle=':')
```

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x,y,linestyle='-.')#더 많은건 matplotlib linestyle검색해서 찾아보렴
```

```python
plt.plot(x,y,linestyle='-')#linestyle안써도 됨(기본ㄴ값)
```

```python
plt.plot(x,y,color='pink')
```

```python
plt.plot(x,y,color='#ff0000')
```

```python
plt.plot(x,y,color='b')
```

```python
plt.plot(x,y,color='g')#matplotlib color에서 미리 지정한 기본 컬러(한글자만 넣어도됨)
```

## Format

```python
plt.plot(x,y,'ro--')#color,marker,linestyle(순서대로)
```

```python
plt.plot(x,y,'bv:')
```

```python
plt.plot(x,y,'go')#아무것도 안적으면 None처리됨
```

## 축약어

```python
plt.plot(x,y,marker='o',mfc='red',ms=10,mec='blue',ls=':')
```

## 투명도

```python
plt.plot(x,y,marker='o',mfc='red',alpha=0.3)#alpha: 투명도(0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=200)#dpi:dots per inch, 확대
plt.plot(x,y)
```

## Background Color

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x,y)
```
