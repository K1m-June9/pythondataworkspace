# 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='AppleGothic'#Mac
#matplotlib.rcParams['font.size']=10 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False#한글폰트 사용 시 -값 깨짐 방지
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
plt.title('HaHa',fontdict={'family':'Helvetica Neue','size':20})#개별 폰트 설정
```

```python
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red')
plt.ylabel('Y축',color='#00aa00')
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red',loc='right')#loc: left,center,right
plt.ylabel('Y축',color='#00aa00',loc='top')#loc:top,center,bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
```
